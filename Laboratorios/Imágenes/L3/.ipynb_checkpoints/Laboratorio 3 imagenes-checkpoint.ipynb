```python
import re
##Función para ordenar las imágenes
def sorted_nicely( l ): 
    """ Sort the given iterable in the way that humans expect.""" 
    convert = lambda text: int(text) if text.isdigit() else text 
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(l, key = alphanum_key)
```

```python
#Función para guardar las imagenes en listas
def cargar_im(carpeta):
    res = {}
    car = sorted_nicely(os.listdir(carpeta))
    for i in range(0,len(car)):
        a = cv.imread(carpeta+'/'+car[i], cv.IMREAD_GRAYSCALE)
        b = car[i]
        res[b] = a
    return res
```

**Implemente un detector de movimiento por diferencia de cuadros**

```python
def detectar_mov(secuencia, umbral, modo = 0):
    """
    Función que recibe una secuencia de imágenes y aplica la diferencia de cuadros.
    Secuencia: Diccionario con la secuencia de imagenes
    umbral: umbral a partir del cual se detecta comop movimientro en la figura.
    modo: 1 muestra todas las figuras (n-1), 0 solo retorna una lista con las diferencias de las matrices
    y la diferencias en binario al aplicar el umbral.

    """ 
    start = time.time()
    names_images = list(secuencia.keys())
    res_binaria = []
    res_dif = []

    for i in range(len(names_images)-1):
        dif = abs(secuencia[names_images[i]].astype(int)-secuencia[names_images[i+1]].astype(int))
        bina = np.where(dif >= umbral, 1, 0)
        res_binaria+=[bina]
        res_dif += [dif]
        if modo == 1:
            print(names_images[i], names_images[i+1])
            plt.imshow(bina, cmap = 'gray', vmin=0, vmax=1)
            plt.show()
    end = time.time()
    print(f'Tiempo de ejecución: {round(end - start,3)} [s].')
    return np.array(res_dif,dtype='uint8'), np.array(res_binaria,dtype='uint8') 
```

**Utilizando la secuencia de imágenes de la carpeta “fondo”, genere un modelo de fondo de
dos matrices calculando el promedio de los cuadros (matriz 1) y la desviación estándar (matriz
2).**

Se toman las imágenes de la carpeta fondo y se obtienen dos matrices, una con el promedio y otra con la std de la carpeta.

```python
def mean_fondo():
    """
    Función que entrega una matriz con el promedio de las imágenes de la carpeta fondo, es decir,
    calcula el promedio para cada pixel de las figuras.
    """
    names_fondo = list(fondo.keys())
    dim = fondo[names_fondo[0]].shape
    res_mean = np.zeros((dim[0],dim[1]))
    for names in names_fondo:
        res_mean += fondo[names] 
        
    return res_mean/len(names_fondo)

mean_fondo = mean_fondo()
```

```python
def std_fondo():
    """
    Función que junta todas las imágenes de la carpeta fondo en un array 3d y luego retorna el array 2d con la std.
    """
    names_fondo = list(fondo.keys())
    dim = fondo[names_fondo[0]].shape
    std = np.empty(shape=(dim[0], dim[1], len(names_fondo)))
    res_std = np.zeros((dim[0],dim[1]))
    for i in range(len(names_fondo)):
        std[:,:,i] = fondo[names_fondo[i]]
    for i in range(dim[0]):
        for j in range(dim[1]):
            res_std[i][j]=np.std(std[i,j,:])
    return res_std
    
    #np.vstack((a,b[None]))
std_fondo = std_fondo()
```

**Elija un umbral apropiado y aplíquelo para detectar movimiento sobre las secuencias de
imágenes.** **Entregue las matrices del modelo de fondo.**

```python
def detectar_mov_fondo(secuencia, alpha, std = std_fondo, mean = mean_fondo):
    limite_sup = mean+alpha*std#matriz
    limite_inf = mean-alpha*std#matriz
    res_mov = []
    names_secuencia = list(secuencia.keys())
    for name in names_secuencia:
        image = secuencia[name]
        bina = np.where((image >= limite_sup) | (image <= limite_inf), 1, 0)
        res_mov+=[bina]
    return np.array(res_mov)
        
```

**Usando la detección de movimiento anterior, calcule el histograma proyectado por columnas (suma de pixeles por columna) y el histograma proyectado por filas (suma de pixeles por fila)**

```python
def get_hist(secuencia_mov):
    res_column = []
    res_fila = []
    for image_mov in secuencia_mov:
        sum_column = image_mov.sum(axis = 0)
        sum_fila = image_mov.sum(axis = 1)
        res_column += [sum_column]
        res_fila += [sum_fila]
    return np.array(res_fila), np.array(res_column)
```

```python
fila_seq1, column_seq1 = get_hist(mov_fondo_seq1)
fila_seq2, column_seq2 = get_hist(mov_fondo_seq2)
fila_seq3, column_seq3 = get_hist(mov_fondo_seq3)
```

**Con esta información, encierre en una caja de dimensiones adecuadas el blob de la detección de movimiento.**

```python
#Para realizar el blob debo conocer el tamaño y posición del mismo

def bordes(H_fila, H_column, modo = 'cent',umbral=8):
    """
    En base al histograma de las filas y columnas, un modo y un umbral.
    """
    #Clasificar si es mayor a umbral como un mov.
    con_fil = np.where(H_fila > umbral, 1, 0)
    con_col = np.where(H_column > umbral, 1, 0)
    #Indices de valores no zero
    fil = np.nonzero(con_fil)[0]
    col = np.nonzero(con_col)[0]
    #print(fil, col)
    if len(fil) == 0 or len(col) == 0: #No detecta mov.
        return 0, 0, 0, 0
    #Limites inferiores y superiores para los histogramas.
    h_ini = fil[0]
    h_fin = fil[-1]
    w_ini = col[0]
    w_fin = col[-1]
    #Ancho, largo, (x,y) el centro.
    H = h_fin - h_ini
    W = w_fin - w_ini
    x = w_ini + W//2
    y = h_ini + H//2
    #Dos distintos modos, uno retorna las esquinas o el centro con el largo y ancho.
    if modo == 'esq':
        return h_ini, h_fin, w_ini, w_fin
    if modo == 'cent':
        return H, W, x, y


def caja(fila, column, imag):
    copy = imag.copy()
    hi, hf, wi, wf = bordes(fila, column, 'esq')
    for i in range(hi, hf):
        copy[i][wi]=150
        copy[i][wf]=150
    for j in range(wi, wf):
        copy[hi][j]=150
        copy[hf][j]=150
    plt.imshow(copy, cmap='gray', vmin=0, vmax=1)
    plt.show()
```

**Utilizando la información de la posición de la caja en el cuadro actual y la de los cuadros anteriores, estime la posición del objeto para el cuadro siguiente. Dibuje la caja estimada y compárela con la obtenida usando el procedimiento descrito en 3. Comente. 

```python
def parametros(seq_hfila, seq_hcolumn, secuencia, umbral=8):
    if secuencia == 1:
        a, b, c = 15, 16, 17
        
    if secuencia == 2:
        a, b, c = 80, 81, 82
        
    if secuencia == 3:
        a, b, c = 80, 81, 82
    
    H_02, W_02, x_02, y_02 = bordes(seq_hfila[a], seq_hcolumn[a], 'cent', umbral=umbral)
    H_01, W_01, x_01, y_01 = bordes(seq_hfila[b], seq_hcolumn[b], 'cent', umbral=umbral)
    H_00, W_00, x_00, y_00 = bordes(seq_hfila[c], seq_hcolumn[c], 'cent', umbral=umbral)
    
    A = round((H_00-H_01)/(H_01-H_02))
    B = round((W_00-W_01)/(W_01-W_02))
    C = round((x_00-x_01)/(x_01-x_02))
    D = round((y_00-y_01)/(y_01-y_02))
    
    return A, B, C, D

def estimacion(seq_hfila, seq_hcolumn, seq, i, secuencia, umbral=8):
    copy = seq[i].copy()
    A, B, C, D = parametros(seq_hfila, seq_hcolumn, secuencia, umbral=umbral)
    H_02, W_02, x_02, y_02 = bordes(seq_hfila[i-2], seq_hcolumn[i-2], umbral=umbral)
    H_01, W_01, x_01, y_01 = bordes(seq_hfila[i-1], seq_hcolumn[i-1], umbral=umbral)

    H_00 = round(H_01 + A*(H_01-H_02))
    W_00 = round(W_01 + B*(W_01-W_02))
    x_00 = round(x_01 + C*(x_01-x_02))
    y_00 = round(y_01 + D*(y_01-y_02))
    
    
    wi = x_00 - W_00//2
    wf = x_00 + W_00//2
    hi = y_00 - H_00//2
    hf = y_00 + H_00//2
    
    
    for i in range(hi, hf):
        
        
        copy[i][wi]=150
        copy[i][wf]=150
    for j in range(wi, wf):
        if hf>799:
            hf = 799
        copy[hi][j]=150
        copy[hf][j]=150
    plt.imshow(copy, cmap='gray', vmin=0, vmax=1)
    plt.show()
    
    
```

```python
for i in range(2,18,3):
    estimacion(fila_seq1, column_seq1, mov_fondo_seq1, i, 1, umbral=8)
```

```python
for i in range(70,110,8):
    estimacion(fila_seq2, column_seq2, mov_fondo_seq2, i, 2, umbral=8)
```

```python
for i in range(70,110,8):
    estimacion(fila_seq3, column_seq3, mov_fondo_seq3, i, 3, umbral=8)
```
