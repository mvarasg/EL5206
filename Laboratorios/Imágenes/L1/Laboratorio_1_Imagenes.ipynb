```python
balls=plt.imread("balls.jpg") 
plt.imshow(balls)
```

```python
#Obtener los 3 canales de color
R = balls[:,:,0]
G = balls[:,:,1]
B = balls[:,:,2]
#Fórmula para convertir en escala de grises
balls_gray = 0.2989 * R + 0.5870 * G + 0.1140 * B

plt.imshow(balls_gray, cmap='gray')
```

```python
#Mostrar histograma, el método .ravel hace un reshape de la imagen a 1D
plt.hist(balls_gray.ravel(), bins=256)
```

```python
def binaria(t,img,color='gray'):
    aux = []
    for i in img.ravel():
        if i <= t:
            aux += [0]
        else:
            aux += [1]
    aux = np.array(aux)
    aux = np.reshape(aux,(640,640))
    #return aux
    plt.title(f'Imagen binaria con umbral t = {t}.')
    plt.imshow(aux, cmap=color)
    plt.show()
```

```python
binaria(128,balls_gray)
```

```python
t = [25, 50, 75, 80, 100, 125, 128, 150, 175]
for i in t:
    
    binaria(i,balls_gray)
```

```python
plt.hist(G.ravel(), bins = 256)
```

```python
g = [150, 170, 190, 210, 230]
for i in g:
    binaria(i,G,'Greens')
```

```python
plt.hist(B.ravel())
```

```python
for i in t:
    binaria(i,B,'Blues')
```

- **Diseñe una transformación mediante un estiramiento lineal de contraste que mejore el contraste en las imágenes.**

La función a utilizar es la siguiente: 

```python
def estiramientoLineal(image):
    return (image-np.min(image))/(np.max(image)-np.min(image))*255
#((píxel - mínimo) / (máximo - mínimo)) * 255.
    
```

- **Indique la transformación (ecuación) y aplíquela a la imagen original.**

```python
moon_est = estiramientoLineal(moon)
old_est = estiramientoLineal(old)
river_est = estiramientoLineal(river)
landscape_est = estiramientoLineal(landscape)
```

- **Calcule la Transformada de Fourier en 2D (en escala de grises) de la imagen ‘sunset.jpg’ usando el método ‘fft2’.**

```python
sunset = plt.imread('sunset.jpg')
```

```python
R_sunset = sunset[:,:,0]
G_sunset = sunset[:,:,1]
B_sunset = sunset[:,:,2]
#Fórmula para convertir en escala de grises
sunset_gray = 0.2989 * R_sunset + 0.5870 * G_sunset + 0.1140 * B_sunset
plt.imshow(sunset_gray, cmap='gray', vmin=0, vmax=255)
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(abs(np.fft.fftshift(np.fft.fft2(sunset_gray))))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.show()
```

- **Visualice su magnitud y comente lo que observa.**

- **Para corregir la visualización calcule la transformación logarítmica de la transformada de Fourier descrita en la ecuación. ¿Por qué se produce el cambio en la forma en que se visualiza la transformada?**

```python
def TLogarit(transformada):
    Rr = abs(np.amax(transformada))
    c = 255/np.log(1+Rr)
    resultado = c*np.log(1+(transformada))
    return resultado
```

```python
plt.figure(figsize=(15, 12))
plt.plot(TLogarit(abs(np.fft.fftshift(np.fft.fft2(sunset_gray)))))
plt.show()
```
