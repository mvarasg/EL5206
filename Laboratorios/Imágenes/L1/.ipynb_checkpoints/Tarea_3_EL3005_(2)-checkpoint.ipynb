## Parte a)
#### Entregar una tabla con los coeficientes del filtro pere los casos de un recorte con ventana rectangular y ventana de Hamming

Para calcular los coeficientes del filtro se utiliza el método de la serie de Fourier para aproximar filtros FIR, lo que se traduce en la fórmula: $$c_{m} = \int_{0}^{1} A_{d}(\nu) cos(m \pi \nu) d \nu$$
con $\nu = \frac{f}{f_{o}} = \frac{f}{100}$, ya que $f_{o} = \frac{f_{s}}{2} = 100 Hz$ y $$A_{d}(f)= \left\{
\begin{array}{ll}
1 & 11 Hz \leq f \leq 16 Hz  \\
0 & \mbox{cualquier otro caso}
\end{array}
\right.
$$  
Dado esto, se puede obtener $c_{m}$ según: $$c_{m} = \int_{0.11}^{0.16} cos(m \pi \nu) d \nu$$

Luego, para aplicar cada ventana se usará $c'_{m} = w_{m} \cdot c_{m}$, donde los $w_{m}$ corresponden a los pesos de cada ventana.

Considerando que la ventana rectángular posee $w_{m}=1$ se tiene que $c'_{m}=c_{m}$. Por lo que los coeficientes correspondientes a la ventana rectángular se encuentran en la siguiente tabla: 

| $c_{m}$ | valor | $c_{m}$ | valor | $c_{m}$ | valor |
|:-------:|:-----:|:-------:|:-----:|:-------:|:-----:|
| $c_{0}$ | 0.05  | $c_{7}$ |-0.0468|$c_{14}$ |0.0381 |
| $c_{1}$ |0.0455 | $c_{8}$ |-0.0453|$c_{15}$ |0.0391 |
| $c_{2}$ |0.0329 | $c_{9}$ |-0.0359|$c_{16}$ |0.0332 |
| $c_{3}$ |0.0146 |$c_{10}$ |-0.0204|$c_{17}$ |0.0219 |
| $c_{4}$ |-0.0062|$c_{11}$ |-0.0021|$c_{18}$ |0.0076 |
| $c_{5}$ |-0.0255|$c_{12}$ |0.0158 |$c_{19}$ |-0.0068|
| $c_{6}$ |-0.0398|$c_{13}$ |  0.03 |$c_{20}$ |-0.0187|

Para obtener los valores de los coeficientes utilizando una ventana de Hamming se multiplican los valores anteriormente obtenidos por su peso respectivo, el cual viene dado por: $$w_{m} = 0.54 + 0.46 \cdot cos \left( \frac{2 \pi m}{40} \right)$$ , ya que 40 es el número de retardos que se utiliza.  A partir de lo anterior, se obtienen los coeficientes indicados en la siguiente tabla:

| $c'_{m}$ | valor | $c'_{m}$ | valor | $c'_{m}$ | valor |
|:--------:|:-----:|:--------:|:-----:|:--------:|:-----:|
| $c'_{0}$ | 0.05  | $c'_{7}$ |-0.035 |$c'_{14}$ |0.0102 |
| $c'_{1}$ |0.0452 | $c'_{8}$ |-0.0309|$c'_{15}$ |0.0084 |
| $c'_{2}$ |0.0322 | $c'_{9}$ |-0.0219|$c'_{16}$ |0.0056 |
| $c'_{3}$ |0.0128 |$c'_{10}$ |-0.011 |$c'_{17}$ |0.0028 |
| $c'_{4}$ |-0.0056|$c'_{11}$ |-0.0097|$c'_{18}$ |0.0008 |
| $c'_{5}$ |-0.022 |$c'_{12}$ |0.0063 |$c'_{19}$ |-0.0006|
| $c'_{6}$ |-0.0322|$c'_{13}$ |0.0099 |$c'_{20}$ |-0.0015|

Se puede notar que los cambios van siendo más significativos conforme avanza el índice, además cabe notar que $c'_{0}$ se mantuvo igual, puesto que le corresponde $w_{0}=1$ lo cual está bien, ya que el lóbulo de mayor tamaño se encuentra centrado.

## Parte b)
#### Aplicar los dos filtros a las siguientes señales:

Se deben convolucionar la señal de entrada con el filtro respectivo 

```python
#Señales
def Señali(t):
    return sin(2*pi*5*t)

def Señalii(t):
    return sin(2*pi*9*t)

def Señaliii(t):
    return sin(2*pi*13*t)

def Señaliv(t):
    return sin(2*pi*20*t)

def Señalv(t):
    return sin(2*pi*25*t)

def Señalvi(t):
    return sin(2*pi*30*t)

#Frecuencia de muestreo
f=200   #Hz

#Ventana temporal del muestreo
ti=0 #seg
tf=1 #seg

#eje temporal:
t=linspace(ti, tf, f)

#Muestreo de las señales:
#Señal i
si=np.zeros(len(t))
for i in range(0,len(t)):
    si[i]+=Señali(t[i])
#Señal ii
sii=np.zeros(len(t))
for i in range(0,len(t)):
    sii[i]+=Señalii(t[i])    
#Señal iii
siii=np.zeros(len(t))
for i in range(0,len(t)):
    siii[i]+=Señaliii(t[i])
#Señal iv
siv=np.zeros(len(t))
for i in range(0,len(t)):
    siv[i]+=Señaliv(t[i])    
#Señal v
sv=np.zeros(len(t))
for i in range(0,len(t)):
    sv[i]+=Señalv(t[i])
#Señal vi
svi=np.zeros(len(t))
for i in range(0,len(t)):
    svi[i]+=Señalvi(t[i])
```

```python
from scipy import signal

Frect = np.array([-0.0187, -0.0068, 0.0076, 0.0219, 0.0332, 0.0391, 0.0381, 0.03, 0.0158, -0.0021, -0.0204, -0.0359, -0.0453, -0.0468, -0.0398, -0.0255, -0.0062, 0.0146, 0.0329, 0.0455, 0.05, 0.0455, 0.0329, 0.0146, -0.0062, -0.0255, -0.0398, -0.0468, -0.0453, -0.0359, -0.0204, -0.0021, 0.0158, 0.03, 0.0381, 0.0391, 0.0332, 0.0219, 0.0076, -0.0068, -0.0187])

FHamm = np.array([-0.0015, -0.0006, 0.0008, 0.0028, 0.0056, 0.0084, 0.0102, 0.0099, 0.0063, -0.0097, -0.011, -0.0219, -0.0309, -0.035, -0.0322, -0.022, -0.0056, 0.0128, 0.0322, 0.0452, 0.05, 0.0452, 0.0322, 0.0128, -0.0056, -0.022, -0.0322, -0.035, -0.0309, -0.0219, -0.011, -0.0097, 0.0063, 0.0099, 0.0102, 0.0084, 0.0056, 0.0028, 0.0008, -0.0006, -0.0015])


#Aplicación de filtro con ventana rectángular
outRi = signal.fftconvolve(si, Frect, mode='same')
outRii = signal.fftconvolve(sii, Frect, mode='same')
outRiii = signal.fftconvolve(siii, Frect, mode='same')
outRiv = signal.fftconvolve(siv, Frect, mode='same')
outRv = signal.fftconvolve(sv, Frect, mode='same')
outRvi = signal.fftconvolve(svi, Frect, mode='same')

#Aplicación de filtro con ventana Hamming
outHi = signal.fftconvolve(si, FHamm, mode='same')
outHii = signal.fftconvolve(sii, FHamm, mode='same')
outHiii = signal.fftconvolve(siii, FHamm, mode='same')
outHiv = signal.fftconvolve(siv, FHamm, mode='same')
outHv = signal.fftconvolve(sv, FHamm, mode='same')
outHvi = signal.fftconvolve(svi, FHamm, mode='same')
```

```python
#Gráficos de la señal de entrada y salida en los casos (i), (iii) y (vi)

#Ventana rectangular
plt.figure(figsize=(12,4))                            #Señal i
plt.plot(t, si,linewidth=0.5)
plt.plot(t, outRi,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana rectángular para una señal sinusoidal de 5 Hz')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #Señal iii
plt.plot(t, siii,linewidth=0.5)
plt.plot(t, outRiii,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana rectángular para una señal sinusoidal de 13 Hz')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #Señal vi
plt.plot(t, svi,linewidth=0.5)
plt.plot(t, outRvi,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana rectángular para una señal sinusoidal de 30 Hz')
plt.legend(fontsize=8)

#Ventana de Hamming
plt.figure(figsize=(12,4))                            #Señal i
plt.plot(t, si,linewidth=0.5)
plt.plot(t, outHi,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana de Hamming para una señal sinusoidal de 5 Hz')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #Señal iii
plt.plot(t, siii,linewidth=0.5)
plt.plot(t, outHiii,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana de Hamming para una señal sinusoidal de 13 Hz')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #Señal vi
plt.plot(t, svi,linewidth=0.5)
plt.plot(t, outHvi,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana de Hamming para una señal sinusoidal de 30 Hz')
plt.legend(fontsize=8)
```

De los gráficos obtenidos se puede observar que, como es de esperar, las amplitudes de las señales (i) y (vi), de $5 [Hz]$ y $30 [Hz]$ respectivamente, se ve disminuida drásticamente, debido a que al ser señales de un frecuencia pura que no se encuentra en el rango del filtro pasabandas creado, son filtradas; mientras que en el caso de la señal (iii) de $13 [Hz]$ se observa que su amplitud no tiene grandes cambios, siendo estos más signficativos al utilizar la ventana de Hamming. 

```python
#Calculo de la relación entre las amplitudes máximas (señalde salida vs de entrada)

#La siguiente función sirve para calcular la relación buscada

def relacion(ent,sal):
    E = max(ent) #Encuentra amplitud máxima de la señal de entrada
    S = max(sal) #Encuentra amplitud máxima de la señal de salida
    r = (S/E) #Representa la relación buscada, es la cantidad retornada por la función
    return r

#En lo que sigue se define la relación A_max-in/A_max-out para cada señal y para cada tipo de ventana

#Señal i rectángular 
RRi = relacion(si,outRi)
#Señal i de Hamming
RHi = relacion(si,outHi)

#Señal ii rectángular 
RRii = relacion(sii,outRii)
#Señal ii de Hamming
RHii = relacion(sii,outHii)

#Señal iii rectángular 
RRiii = relacion(siii,outRiii)
#Señal iii de Hamming
RHiii = relacion(siii,outHiii)

#Señal iv rectángular 
RRiv = relacion(siv,outRiv)
#Señal iv de Hamming
RHiv = relacion(siv,outHiv)

#Señal v rectángular
RRv = relacion(sv,outRv)
#Señal v de Hamming
RHv = relacion(sv,outHv)

#Señal vi rectángular
RRvi = relacion(svi,outRvi)
#Señal vi de Hamming
RHvi = relacion(svi,outHvi)

#Se definen vectores que contienen las relaciones para cada señal
Rrect=np.array([RRi, RRii, RRiii, RRiv, RRv, RRvi]) #Ventana rectangular
Rhamming=np.array([RHi, RHii, RHiii, RHiv, RHv, RHvi]) #Ventana de Hamming
```

```python
#Gráficos relación calculada vs frecuencia:

#Se define un vector que contiene las frecuencias de las señales
frec=np.array([5,9,13,20,25,30])

#Se grafica para ambas ventanas
plt.figure(figsize=(12,4))                            #ventana rectangular
plt.plot(frec, Rrect,'*')
plt.plot(frec, Rrect,linewidth=0.5)
plt.xlabel('frecuencia de la señal de entrada [Hz]', fontsize=14)
plt.ylabel('Relación $A_{max-out}/A_{max-in}$', fontsize=14)
plt.title('Relación amplitudes máximas vs frecuencias de las señales de entrada [ventana rectangular]')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #ventana de Hamming
plt.plot(frec, Rhamming,'*')
plt.plot(frec, Rhamming,linewidth=0.5)
plt.xlabel('frecuencia de la señal de entrada [Hz]', fontsize=14)
plt.ylabel('Relación $A_{max-out}/A_{max-in}$', fontsize=14)
plt.title('Relación amplitudes máximas vs frecuencias de las señales de entrada [ventana de Hamming]')
plt.legend(fontsize=8)
```

## Parte c)
#### Repetir las partes (a) y (b) utilizando 600 retardos

```python
#Cálculo de coeficientes utilizando 600 retardos

#La función coef realiza el cálculo de los c_m expresados en la parte (a)
def coef(m):    
    return (sin(m*pi*0.16)-sin(m*pi*0.11))/(m*pi)

#La función pesoHamming realiza el cálculo de cada uno de los pesos para la ventana de Hamming
def pesoHamming(m):
    return 0.54 + 0.46*cos(2*pi*m/40)

#La función coefHamming entrega el coeficiente final luego de multiplicar por el peso correspondiente
def coefHamming(m):
    return coef(m)*pesoHamming(m)

#Ventana rectángular
#La funcionen coeficientes y coeficientesHamm entregan arreglos que contienen los coeficientes para cada ventana

def coeficientes(retardos):
    C = np.zeros(retardos//2 + 1)
    C[0]+=0.05
    for i in range(1, len(C)):
        C[i]+=coef(i)
    return C

#Ventana de Hamming
def coeficientesHamm(retardos):
    C = np.zeros(retardos//2 + 1)
    C+=0.05
    for i in range(1, len(C)):
        C[i]+=coefHamming(i)
    return C

M = linspace(0, 300, 301)

plt.figure(figsize=(12,4))                            #ventana rectangular
plt.plot(M, coeficientes(600),linewidth=0.5)
plt.plot(M, coeficientesHamm(600),linewidth=0.5)
plt.xlabel('índice del coeficiente', fontsize=14)
plt.ylabel('Valor coeficientes', fontsize=14)
plt.title('Visualización de los coeficientes para cada ventana')
plt.legend(fontsize=8)     
```

En el gráfico anterior se observa cómo se distribuyen los valores tomados por los coeficientes para cada una de las ventanas, donde, la curva azul corresponde a los coeficientes con ventana rectangular y la naranja a los coeficientes con ventana de Hamming, si bien ambas distribuciones son similares, los coeficientes de Hamming son mayores que los de la ventana rectangular.

```python
#La función relleno se encarga de rellenar un arreglo con los coeficientes del filtro de manera simétrica
#respecto a la mitad de la cantidad de retardos +1

def relleno(C,retardos):
    F = np.zeros(retardos + 1)
    F[retardos//2+1]+=C[0]
    for i in range(1,retardos//2 +1):
        F[retardos//2 -i]+=C[i]
        F[retardos//2 +i]+=C[i]
    return F

FRECT= relleno(coeficientes(600), 600)
FHAMM= relleno(coeficientesHamm(600),600)

#Aplicación de filtro con ventana rectángular
OutRi = signal.fftconvolve(si, FRECT, mode='same')
OutRii = signal.fftconvolve(sii, FRECT, mode='same')
OutRiii = signal.fftconvolve(siii, FRECT, mode='same')
OutRiv = signal.fftconvolve(siv, FRECT, mode='same')
OutRv = signal.fftconvolve(sv, FRECT, mode='same')
OutRvi = signal.fftconvolve(svi, FRECT, mode='same')

#Aplicación de filtro con ventana Hamming
OutHi = signal.fftconvolve(si, FHAMM, mode='same')
OutHii = signal.fftconvolve(sii, FHAMM, mode='same')
OutHiii = signal.fftconvolve(siii, FHAMM, mode='same')
OutHiv = signal.fftconvolve(siv, FHAMM, mode='same')
OutHv = signal.fftconvolve(sv, FHAMM, mode='same')
OutHvi = signal.fftconvolve(svi, FHAMM, mode='same')
```

```python
#Gráficos de entrada y salida de ambos filtros en las señales (i), (iii) y (vi)

#Ventana rectangular
plt.figure(figsize=(12,4))                            #Señal i
plt.plot(t, si,linewidth=0.5)
plt.plot(t, OutRi,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana rectángular para una señal sinusoidal de 5 Hz')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #Señal iii
plt.plot(t, siii,linewidth=0.5)
plt.plot(t, OutRiii,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana rectángular para una señal sinusoidal de 13 Hz')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #Señal vi
plt.plot(t, svi,linewidth=0.5)
plt.plot(t, OutRvi,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana rectángular para una señal sinusoidal de 30 Hz')
plt.legend(fontsize=8)

#Ventana de Hamming
plt.figure(figsize=(12,4))                            #Señal i
plt.plot(t, si,linewidth=0.5)
plt.plot(t, OutHi,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana de Hamming para una señal sinusoidal de 5 Hz')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #Señal iii
plt.plot(t, siii,linewidth=0.5)
plt.plot(t, OutHiii,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana de Hamming para una señal sinusoidal de 13 Hz')
plt.legend(fontsize=8)

plt.figure(figsize=(12,4))                            #Señal vi
plt.plot(t, svi,linewidth=0.5)
plt.plot(t, OutHvi,linewidth=0.5)
plt.xlabel('tiempo [s]', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title('Entrada y salida del filtro con ventana de Hamming para una señal sinusoidal de 30 Hz')
plt.legend(fontsize=8)
```
