### Clase implementada en la auxiliar.

```python
import dss

class DSS(object):
    def __init__(self,dssFileName):
        
        #Creamos una nueva instancia de Opendss
        self.dssObj = dss.DSS
        
        #asignamos variables para acceder mas rápido
        self.dssText = self.dssObj.Text
        self.dssCircuit = self.dssObj.ActiveCircuit
        self.dssSolution = self.dssCircuit.Solution
        self.dssMeters = self.dssCircuit.Meters
        self.dssMonitors = self.dssCircuit.Monitors
        self.dssCktElement = self.dssCircuit.ActiveCktElement
        self.dssFileName = dssFileName
        self.power_monitors = []
        self.voltage_monitors = []
        #Limpiamos DSS para cargar un circuito nuevo
        self.dssObj.ClearAll()
        
    def run_command(self, text):
        #Para escribir un comando directo en opendss
        self.dssText.Command = text
    
    def Compile(self):
        #Compilamos el archivo 
        self.run_command("Compile "+ self.dssFileName)
        
    def mySolve(self):
                self.dssSolution.Mode = 1 #equivale a mode=daily
                self.dssSolution.ControlMode = 0 #equivale a controlmode=static
                self.dssSolution.Number = 1440
                self.dssSolution.StepsizeMin = 1
                self.dssSolution.Solve()   
                
        
    def ShowVoltages(self):
        self.run_command("Show voltages")
        
    #MONITORES

    def add_monitor(self, atype: str, element: str, polar: bool, terminal: int):
        
        #funcion para agregar un monitor a un elemento
        split_element = element.split('.')
        assert len(split_element) == 2
        if atype == 'voltaje': mode =0
        elif atype == 'potencia': mode= 1
        elif atype == 'perdidas': mode=9   
        
        if polar==True: polar='yes'
        else: polar='no'
        monitor_name = f'{split_element[1]}_{atype}_t{terminal}'.lower()
        self.run_command( f'new monitor.{monitor_name} element={element} mode={mode} ' \
                                       f'terminal={terminal} ppolar={polar}')
        #podemos guardar los monitores
        if atype == 'potencia':
                self.power_monitors.append(monitor_name)
        elif atype == 'voltaje':
                self.voltage_monitors.append(monitor_name) 
                
    def export_monitors(self, directory_save:str, atype:str,element:str,channels:list,terminal:int):
        
        #funcion para exportar todos los monitores
        split_element= element.split('.')
        assert len(split_element) == 2
        if atype == 'voltaje': mode = 0
        elif atype == 'potencia': mode = 1
        elif atype == 'perdidas': mode = 9
            
        #definimos nombre automatizado del monitor
        monitor_name= f'{split_element[1]}_{atype}_t{terminal}'.lower()
        
        #obtenemos el header del monitor (para no exportar canales que no necesitamos y ahorrar espacio)
        DSSMonitors = self.dssMonitors 
        DSSMonitors.Name = monitor_name #activamos el monitor que queremos
        header=DSSMonitors.Header #accedemos a los nombres del header del monitor
        
        if channels==[]:
            channels=header
        #por cada canal en la lista de canales se agrega una columna en datos de df       
        monitor_data = {}
        for channel in channels:
            index = header.index(channel)+1
            monitor_data[channel] = DSSMonitors.Channel(index) 
            
        #exportamos resultados
        df = pd.DataFrame(monitor_data)
        df.to_csv(directory_save+'\\'+self.dssCircuit.Name+'_Mon_'+monitor_name+'.csv')

 

    #ENERGYMETER
    def add_meter(self, element: str, terminal: int):
        """
        It measures not only power and energy values at its location, but losses and overload
        values within a defined region of the circuit. The EnergyMeter object uses the concept of a zone. 
        A zone is a collection of circuit elements "downline" from the meter
        
        Element= Name of an existing circuit element to which the monitor is to be connected
        Terminal= No. of the terminal to which the monitor will be connected, normally the source end.     
        """
        split_element = element.split('.')
        assert len(split_element) == 2
        meter_name = '{}_t{}'.format(element.split('.')[1], terminal).lower() #nombre automatico para el energymeter
        metered_elements = list()
        meter_i = self.dssCircuit.Meters.First
        
        while meter_i:
            metered_elements.append(self.dssCircuit.Meters.MeteredElement)
            meter_i = self.dssCircuit.Meters.Next
        if meter_name in self.dssCircuit.Meters.AllNames:
            print('{meter_name} ya existe')
        elif element.lower() in metered_elements:
            print('Elemento {element} ya está siendo monitoreado por otro medidor')
        else:
            self.run_command('New EnergyMeter.'+str(meter_name)+ ' Element='+str(element)+' Losses=Yes Terminal='+str(terminal))
 
    def export_meters(self,directory_save:str, element:str,terminal:int=1) -> None:
        split_element = element.split('.')
        assert len(split_element) == 2
        meter_name = '{}_t{}'.format(element.split('.')[1], terminal).lower()
        self.run_command('Export Meters '+directory_save+'\\'+str(meter_name)) #con esta línea estamos exportando todos los canales del energymeter seleccionado.
        
```

Función para generar los perfiles de carga de forma aleatoria. Como ya se generaron se comenta la línea que los genera.

```python
def getPerfilCarga(data):
    """
    Retorna la carga dado la cap del transformador. Si la cap del transformador es mayor a cero, divide este valor por 2.0 y toma la parte entera.
    Este número es la cantidad de perfiles que se tomarán de forma aleatoria y se sumarán. Finalmente, se divide en 1000 para que esté en kilos.
    """
    arr = data['Capacidad Transformador [KVA]'].to_numpy()
    c = 0
    for i in range(len(arr)):
        val = arr[i]
        if val >0:
            c+=1
            div = int(val/2.0)
            ind = np.random.randint(0,1999,div)
            df = pd.DataFrame()
            df["sum"] = perfiles_CREST[ind].sum(axis=1)
            df.to_csv(f'Perfil_Carga_{i}.txt', index = False, header = False)
    print(f'En total se procesaron {c} perfiles de carga.')
#getPerfilCarga(Nodos)
```
