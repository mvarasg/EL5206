## Parte 1.1

```python
#Esta funcion se encarga de seccionar la imagen

def division_image(photo):
  '''
  Esta función se encarga de seccionar una imagen de entrada en 5 zonas no 
  rectangulares

  Inputs:
  photo: La foto ya leida, debe estar en HSV

  Outputs:
  (resultado1.astype(np.uint8), c1): Máscara de la zona en cuestión y el área de esta (Zona de arriba a la izquierda)
  (resultado2.astype(np.uint8), c2): Máscara de la zona en cuestión y el área de esta (Zona de arriba a la derecha)
  (resultado3.astype(np.uint8), c3): Máscara de la zona en cuestión y el área de esta (Zona de abajo a la derecha)
  (resultado4.astype(np.uint8), c4): Máscara de la zona en cuestión y el área de esta (Zona de abajo a la izquierda)
  (resultado5.astype(np.uint8), c5): Máscara de la zona en cuestión y el área de esta (Zona circular ubicada en el centro)
  '''
  x_photo = photo.shape[1] #Ancho de la foto
  y_photo = photo.shape[0] #Alto de la foto
  z_photo = photo.shape[2] #profunidad de la foto
  #Contadores de área
  c1 = 0
  c2 = 0
  c3 = 0
  c4 = 0
  c5 = 0

  if z_photo == 3:
    resultado1 = np.zeros([y_photo, x_photo, 1])
    resultado2 = np.zeros([y_photo, x_photo, 1])
    resultado3 = np.zeros([y_photo, x_photo, 1])
    resultado4 = np.zeros([y_photo, x_photo, 1])
    resultado5 = np.zeros([y_photo, x_photo, 1])
    x_medio = x_photo//2
    y_medio = y_photo//2
    a = (x_medio - x_photo//10)
    b = (y_medio - y_photo//10)
    for i in range(x_photo):
      for j in range(y_photo):
        #Definimos el circulo del area 5
        elipse = ((i-x_medio)**(2)/(a**2)) + ((j-y_medio)**(2)/(b**2))
        if elipse <= 1:
          resultado5[j][i] = 255
          c5 += 1
        if elipse > 1:
          if i <= x_medio and j <= y_medio:
            resultado1[j][i] = 255
            c1 +=1
          elif i > x_medio and j <= y_medio:
            resultado2[j][i] = 255
            c2 +=1
          elif i > x_medio and j > y_medio:
            resultado3[j][i] = 255
            c3 +=1
          elif i <= x_medio and j > y_medio:
            resultado4[j][i] = 255
            c4 +=1
  return ((resultado1.astype(np.uint8), c1), (resultado2.astype(np.uint8),c2), (resultado3.astype(np.uint8), c3), (resultado4.astype(np.uint8), c4), (resultado5.astype(np.uint8), c5))
```

```python
def calculate_histogram(photo, mask):
  '''
  Función encargada de calcular el histograma de una zona en particular

  Inputs:
  photo: Imagen en HSV 
  mask: Matriz que indica la región a trabajar

  Outputs:
  result_1d: Histograma normalizado y en 1 dimensión
  '''
  result = cv2.calcHist([photo], [0,1,2], mask, [8,12,3], [0, 180,0, 255,0,255])
  result_n = cv2.normalize(result, result, 0, 255, cv2.NORM_MINMAX)
  result_1d = result_n.flatten()
  return result_1d
```

## Parte 1.2


```python
def LBP_sintraslape(foto_LBP, ancho, alto):
  '''
  La función LBP_sintraslape extrae un vector de características a partir
  de los histogramas extraidos de la división de la imagen en regiones 
  de tamaño anchoxalto sin traslape.

  Inputs:
  foto_LBP: imagen con LBP aplicado
  ancho: ancho de la región que divide la imagen de entrada
  alto: alto de la región que divide la imagen de entrada

  Outputs:
  histogramas: concatenación de los histogramas de cada región
  '''
  histogramas = np.array([])
  for i in range(foto_LBP.shape[0]//ancho): #filas
    for j in range(foto_LBP.shape[1]//alto): #columnas
      histogram, bin_edges = np.histogram(foto_LBP[i*alto:(i+1)*alto][j*ancho:(j+1)*ancho], bins=59)
      histogramas = np.concatenate((histogramas, histogram), axis=0)
  return histogramas
```

## Función de extracción de características completa.

```python
def feature_vector(image_path, method):
  '''
  Esta función se encarga de calcular el vector de característica de la foto deseada

  Inputs:
  image_path: Path de la imagen deseada
  method: Método de cálculo del vector de características ('color', 'LBP', 'inception', 'mobile')

  Outputs:
  features: Vector de características
  '''
  #Método basado en color (parte 1.1)
  if method == 'color':
    photo = cv2.imread(image_path)
    photo_HSV = cv2.cvtColor(photo, cv2.COLOR_BGR2HSV)
    division_photo = division_image(photo_HSV)
    features = []
    for i in range(len(division_photo)):
      hist_sec = calculate_histogram(photo_HSV, division_photo[i][0])
      features.extend(hist_sec)

  #Método basado en lobal binnary pattern (parte 1.2)
  elif method == 'LBP':
    photo = cv2.imread(image_path)
    photo = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)
    photo = cv2.resize(photo, dsize=(2448//2, 3264//2)) 
    LBP = local_binary_pattern(photo, P = 8, R = 1)
    features = LBP_sintraslape(LBP, 2448//16, 3264//16)

  #Método basado en el uso de la red InceptionResNetV2 (parte 1.3)
  elif method == 'inception':
    base_model = InceptionResNetV2(weights='imagenet', include_top=True)
    model = Model(inputs=base_model.input,outputs=base_model.get_layer('avg_pool').output)

    img = image.load_img(image_path, target_size=(299, 299))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # Get the prediction.
    features = model.predict(x)    
    features = features[0].tolist()

  #Método basado en el uso de la red MobileNetV2 (parte 1.4)
  elif method == 'mobile':
    mobile = tf.keras.applications.MobileNetV2(weights="imagenet", include_top=True)

    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    preparada = tf.keras.applications.mobilenet_v2.preprocess_input(img_array_expanded_dims)

    features = mobile.predict(preparada)
    features = features[0].tolist()
    
  return features
```

## Función de extracción de características óptimizada

```python
#Función de cálculo de vectores de características óptimizada 

def feature_vector_opt(image_path, method):
  '''
  Esta función se encarga de calcular el vector de característica de la foto deseada

  Inputs:
  image_path: Path de la imagen deseada
  method: Método de cálculo del vector de características ('color', 'LBP')

  Outputs:
  features: Vector de características
  '''
  #Método basado en color (parte 1.1)
  if method == 'color':
    photo = cv2.imread(image_path)
    photo_HSV = cv2.cvtColor(photo, cv2.COLOR_BGR2HSV)
    photo_HSV = cv2.resize(photo_HSV, dsize=(2448//4, 3264//4)) 
    division_photo = division_image(photo_HSV)
    features = []
    for i in range(len(division_photo)):
      hist_sec = calculate_histogram(photo_HSV, division_photo[i][0])
      features.extend(hist_sec)

  #Método basado en lobal binnary pattern (parte 1.2)
  elif method == 'LBP':
    photo = cv2.imread(image_path)
    photo = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)
    photo = cv2.resize(photo, dsize=(2448//4, 3264//4)) 
    LBP = local_binary_pattern(photo, P = 8, R = 1)
    features = LBP_sintraslape(LBP, 2448//16, 3264//16)
    
  return features

```

## Parte 2

```python
def distance_images(vector1, vector2, method):
  '''
  Función encargada de calcular la distancia entre 2 vectores de características

  Inputs:
  vector1: Primer vector de características
  vector2: Segundo vector de características
  method: Método para cálcular la distancia ('euclidean', 'cosine')

  Outputs:
  dist: Distancia entre los vectores de características
  '''
  if method == 'euclidean':
    dist = distance.euclidean(vector1, vector2)
  elif method == 'cosine':
    dist = distance.cosine(vector1, vector2)
  return dist
```

```python
#Ranking distancias color euclidean
dist_color_euclidean = pd.read_csv('dist_color.csv', header=None)
dist_color_euclidean = dist_color_euclidean.set_index(0).to_dict()
dist_color_euclidean = dist_color_euclidean[1]

#Ordena las imágenes - Parte 3
keys_color_euclidean = dist_color_euclidean.keys()
rank_color_euclidean = []
rank_norm_color_euclidean = []

for i in keys_color_euclidean:
  orden = sorted(ast.literal_eval(dist_color_euclidean[i]).items(),  key=operator.itemgetter(1))
  dist_color_euclidean[i]=orden

  #Ranking - Parte 4 y 5
  N_rel = 0
  R = 0 
  for j in range(len(orden)):
    if i[:5]==orden[j][0][:5]:
      N_rel += 1
      R += (j+1)
      #print(R)
  rank=R/N_rel
  N = len(orden) 
  rank_norm= (R-(N_rel*(N_rel+1)/2))/(N*N_rel)  
  rank_color_euclidean.append(rank)
  rank_norm_color_euclidean.append(rank_norm)
```

```python
#Promedio y desviación estándar del ranking color euclidean
mean_color_euclidean = np.mean(np.array(rank_color_euclidean))
std_color_euclidean = np.std(np.array(rank_color_euclidean))
print(f'Ranking promedio color euclidean: {mean_color_euclidean}')
print(f'Desviación estándar color euclidean: {std_color_euclidean}')

#Promedio y desviación estándar del ranking normalizado color euclidean
mean_norm_color_euclidean = np.mean(np.array(rank_norm_color_euclidean))
std_norm_color_euclidean = np.std(np.array(rank_norm_color_euclidean))
print(f'Ranking normalizado promedio color euclidean: {mean_norm_color_euclidean}')
print(f'Desviación estándar normalizado color euclidean: {std_norm_color_euclidean}')
```

```python
#Función buscador 

def buscador(img_query, method_feature, method_distance):
  '''
  Esta función recibe una imagen de consulta y expone las 10 imágenes con mayor similitud. 
  Además, muestra el tiempo de consulta, ranking y ranking normalizado.
  
  Inputs:
  img_query: Imagen de consulta terminada en '00', p.j. '134200.jpg'.
  method_feature: Método de extracción de características ('color', 'LBP', 'inception', 'mobile')
  method_distance: Método de cálculo de distancias ('euclidean', 'cosine')
  '''
  inicial = time.time()
  dic_query = {}
  #Path imagen de consulta
  if int(img_query[:2])<13:
    path_query = r'/content/out/out_jpg1/jpg/'+str(img_query)
  else:
    path_query = r'/content/out/out_jpg2/jpg/'+str(img_query)

  #Métodos de extracción de características - Parte 1 y 2
  if method_feature == 'color':
    vector_query = feature_vector(path_query, 'color')
    for j in range(len(img_database)):
      vector_database = feature_color[feature_color.iloc[:,0]==img_database[j]].iloc[0,1:]
      if method_distance == 'euclidean':
        dist = distance_images(vector_query, vector_database.values.tolist(), 'euclidean')
      elif method_distance == 'cosine':
        dist = distance_images(vector_query, vector_database.values.tolist(), 'cosine')
      dic_query[img_database[j]] = dist
    
  elif method_feature == 'LBP':
    vector_query = feature_vector(path_query, 'LBP')
    for j in range(len(img_database)):
      vector_database = feature_LBP[feature_LBP.iloc[:,0]==img_database[j]].iloc[0,1:]
      if method_distance == 'euclidean':
        dist = distance_images(vector_query, vector_database.values.tolist(), 'euclidean')
      elif method_distance == 'cosine':
        dist = distance_images(vector_query, vector_database.values.tolist(), 'cosine')
      dic_query[img_database[j]] = dist
    
  elif method_feature == 'inception':
    vector_query = feature_vector(path_query, 'inception')
    for j in range(len(img_database)):
      vector_database = feature_inception[feature_inception.iloc[:,0]==img_database[j]].iloc[0,1:]
      if method_distance == 'euclidean':
        dist = distance_images(vector_query, vector_database.values.tolist(), 'euclidean')
      elif method_distance == 'cosine':
        dist = distance_images(vector_query, vector_database.values.tolist(), 'cosine')
      dic_query[img_database[j]] = dist

  elif method_feature == 'mobile':
    vector_query = feature_vector(path_query, 'mobile')
    for j in range(len(img_database)):
      vector_database = feature_mobile[feature_mobile.iloc[:,0]==img_database[j]].iloc[0,1:]
      if method_distance == 'euclidean':
        dist = distance_images(vector_query, vector_database.values.tolist(), 'euclidean')
      elif method_distance == 'cosine':
        dist = distance_images(vector_query, vector_database.values.tolist(), 'cosine')
      dic_query[img_database[j]] = dist
  
  #Ordena las imágenes - Parte 3
  orden = sorted(dic_query.items(),  key=operator.itemgetter(1))
  
  #Ranking - Parte 4 y 5
  N_rel = 0
  R = 0 
  for j in range(len(orden)):
    if img_query[:5]==orden[j][0][:5]:
      N_rel += 1
      R += (j+1)
  rank=R/N_rel
  N = len(orden) 
  rank_norm= (R-(N_rel*(N_rel+1)/2))/(N*N_rel)   
  final = time.time()
  tiempo = final - inicial

  #Muestra de resultados - Parte 6
  print(f'Imagen de consulta: {img_query}')
  print(f'Ranking: {rank}')
  print(f'Ranking normalizado: {rank_norm}')
  print(f'Tiempo total de consulta: {tiempo}')
  print('Resultados:')

  lista_orden = orden[:10]
  photo = cv2.imread(path_query)
  #photo = cv2.resize(photo,(0, 0), fx = 0.2, fy = 0.2) 
  #cv2_imshow(photo)
  print('')
  print('')
  print('')

  #Muestra los resultados
  plt.figure()                            #creamos la figura

  fig = plt.figure(figsize=(20, 10))      #Tamaño de la figura, se puede reducir o variar la proporcion cambiando el figsize

  query_img = Image.open(path_query).resize((700,600))
  query_img = np.array(query_img).astype('int32')
  plt.imshow(query_img/255)  
  plt.title(f'Imagen de consulta: {img_query}')
  plt.axis('off')  
  #plt.savefig('query.png')
  plt.show()                      
  
  plt.figure()                            #creamos la figura

  fig1 = plt.figure(figsize=(10, 20))      #Tamaño de la figura, se puede reducir o variar la proporcion cambiando el figsize
  matriz_foto= fig1.add_subplot(5, 2, 1)
  # Plot top 9 retrieval images

  for i in range(len(lista_orden)):
    if int(lista_orden[i][0][:2])<13:
      path_database = r'/content/out/out_jpg1/jpg/'+str(lista_orden[i][0])
    else:
      path_database = r'/content/out/out_jpg2/jpg/'+str(lista_orden[i][0])
    matriz_foto = fig1.add_subplot(5, 2, i+1)    # las posiciones de los retrieval van del 6 al 15
    image = Image.open(path_database).resize((700,600))
    image = np.array(image).astype('int32') #cargamos la imagen
    plt.imshow(image/255)       #mostramos la imagen
    plt.title(f'Posición {i+1}: {lista_orden[i][0]}')
    plt.axis('off')                             #le quitamos los ejes a la imagen
  #plt.savefig('mobilecosine.jpg')              #mostramos la figura
  plt.show()
  plt.close()
  
```
